## Dataframe from Dict of Scalars

```python
summary_dict["Unique Items"] = df["Item ID"].nunique()
summary_dict["Average Price"] = "${0:,.2f}".format(df["Price"].mean()) 
summary_dict["Number of Purhases"] = df["Purchase ID"].count()
summary_dict["Total Revenue"] = "${0:,.2f}".format(df["Price"].sum())
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




Formatting of a DataFrame Column with Map

```python
df_drops = df.drop_duplicates("SN", keep="first")
gender_df = df_drops[["Gender", "Age", "Price"]].groupby("Gender")
gender_counts = gender_df.count()
gender_total = gender_counts.sum()
gender_counts["Price"]= gender_counts/gender_total*100
gender_counts["Price"] = gender_counts["Price"].map('{0:,.2f}%'.format)
gender_counts.columns=["Total Count", "Percentage of Players"]
gender_counts.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age

* Create a summary data frame to hold the results

```python
df_drops2 = df.drop_duplicates("SN", keep="first")
```

```python
df_drops2["Total Count"] = pd.cut(df_drops2['Age'], bins, labels=bin_labels)
```

```python
df_drops2.head()
```

```python
count_items = df_drops2.groupby("Total Count").count()["Item ID"]
price_items = df_drops2.groupby("Total Count").mean()["Price"]
cost_all =  df_drops2.groupby("Total Count").sum()["Price"]
unique_people =  df_drops2.groupby("Total Count").nunique()["SN"]
price_per_person = cost_all/unique_people
purchase_summary_df = pd.DataFrame({"Total # Purchases":count_items, "Total Cost Purchases":cost_all, "Average Price":price_items, "Average Price per Person":price_per_person})
purchase_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = df[['Item ID', 'Item Name', 'Price']]
popular_df.head()
```

```python
popular_grouped_df = popular_df.groupby(['Item ID', 'Item Name'])
popular_counts = popular_grouped_df.count()['Price']
popular_sums = popular_grouped_df.sum()['Price']
popular_mean = popular_grouped_df.mean()['Price']
popular_summary = pd.DataFrame({"Total Items":popular_counts, "Item Price":popular_mean, "Total Purchases":popular_sums})
popular_summary.sort_values(by="Total Items", ascending=False).head()
```
